```python
# CORRELACIÓN
# Gráfico

seed = 13 # seed para reproducibilidad
rand = np.random.RandomState(seed) # Generar números aleatorios
fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(15,8)) #nrows es el número de filas y ncols el de columnas. #figsize es el tamaño de gráfico.

x = rand.uniform (0,1,100) # Generar 100 números aleatorios entre 0 y 1
x = np.vstack((x,x*2+1)) # x es una matriz, donde la fila 0 es x y la fila 1 es 2x+1. #vstack es para apilar verticalmente las filas
x = np.vstack((x,-x[0,]*2+1)) # Agregar una tercera fila, que es -2x+1
x = np.vstack((x,rand.normal(1, 3, 100))) #Agregar una 4ta fila que es una normal con media 1 y desviación estándar 3

for noise, i in zip ([0.05,0.15,0.6,2], [0,1,2,3]): # zip es para iterar sobre dos lienas al mismo tiempo. #los números son el ruido que va agregar cada fila.
    #Añadir ruido. 
    x_with_noise = x+rand.normal(0,noise,x.shape)

    rho_noise = np.corrcoef(x_with_noise) #Calcular correlación entre las filas de x_with_noise

    # Correlación positiva y negativa
    ax[0,i].scatter(x_with_noise[0,],x_with_noise[1], color="blue")
    ax[1,i].scatter(x_with_noise[0,],x_with_noise[2], color="green")
    ax[0,i].title.set_text('Correlación =' + "{:.2f}".format(rho_noise[0,1]) +'/n Ruido =' + "{:.2f}".format(noise))
    ax[1,i].title.set_text('Correlación =' + "{:.2f}".format(rho_noise[0,2]) +'/n Ruido =' + "{:.2f}".format(noise))

    ax[0,i].set(xlabel='x', ylabel='y')
    ax[1,i].set(xlabel='x', ylabel='y')

fig.subplots_adjust(wspace=0.3, hspace=0.4)
plt.show()
```
